```python
import pandas as pd
datos= {"nombre": ["Juan", "Pedro", "maria"],"edad": [20, 30, 40]}
df=pd.DataFrame(datos)
print(df)
```

#OPERACIONES BÁSICAS DE PANDAS

```python
print(df.head()) 
#podemos elegir que filas queremos ver, añadiendo un parametro ,
print("."*20)
print(df.head(2))
#para estadisticas 
print("."*20)
print(df.describe())
print("."*20)
print(df["nombre"])  # acceder a una columna específica

```

#FILTRAR DATOS

```python
print(df["nombre"] == "Juan")  # filtrar filas según una condición, esto genera un booleano---- 0 True, 1 False, 2 False
print("."*20)
print(df[df["nombre"] == "Juan"])  # filtrar por las filas que son TRUE
print("."*20)
print(df.iloc[1])
print("."*20)
print(df.loc[:1, "nombre"])#si fuera iloc[:1] no incluiria la fila1
print("."*20)
print(df.loc[0, ["nombre", "edad"]])
print("."*20)
print(df.loc[df["nombre"]=="Juan", "edad"])  # filtrar por las filas que son TRUE y mostrar la columna edad
```

#MODIFICAR Y AÑADIR DATOS

```python
df["no-niño"] = df["edad"] > 5 
# si no existe la columna "no-niño" la crea y le asigna el valor de la condicion
#df = df.drop(columns=["aprobado"]) 
# eliminar la columna "no-niño"
print(df)
```
