```python
import numpy as np
array = np.array([1,2,3,4,5])
print(array)
print(array[0])
print(type(array))
print(array *2)
```

```python
print(array.sum())
print(array.mean())
print(array.max())
print(array.min())
print(array.shape)
```

```python
matriz= np.array([[1,2,4],
                  [14,6,7]])
print(matriz)
print(matriz.shape)
print(matriz[0])
print(matriz[1,2])
```

#UNIDIMENSIONALES (5,)
#BIDIMENSIONAL (2,3) siendo el primer numero filas y el segundo columnas

#OPERACIONES CON MATRICES:
#AXIS 0 POR COLUMNAS Y AXIS 1 POR FILAS
#Porque el shape(2,4) el primero son las filas y el segundo las columnas
#si es por filas significa "muevete por filas" por tanto recorre las columnas

```python
print(matriz.sum())
print(matriz.sum(axis=0))
print(matriz.sum(axis=1))
print(matriz.mean())
```

#INDEXING Y SLICING

```python
print(matriz[0,1])
print(matriz[:, 1])
print(matriz[1,:])
print(matriz[0,0:3])
print(matriz[0:2, 0])
print(matriz[0:2, 1:3])
```

```python
print(np.eye(3))
print(np.zeros((3,3)))
print(np.ones((3,3)))
print(np.random.rand(1,3))
print(np.random.randint(1,9, size=(3,3)))
```
